```python
import threading
from flask import Flask
import requests
import time

app = Flask(__name__)

def running_in():
    try:
        shell = get_ipython()
        return "Google Colab" if "google.colab" in str(shell) else "Jupyter Notebook"
    except NameError:
        return "a plain Python interpreter"

@app.route("/")
def home():
    return f"<h1>Hello, World!</h1><p>Environment: {running_in()}</p>"

def run_app():
    app.run(port=5000)

thread = threading.Thread(target=run_app)
thread.daemon = True
thread.start()

time.sleep(2)

response = requests.get("http://127.0.0.1:5000")
print(f"This notebook is running in: {running_in()}")
print("Server response:")
print(response.text)
```
